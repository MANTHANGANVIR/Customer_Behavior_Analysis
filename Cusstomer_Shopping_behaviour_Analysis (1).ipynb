```python
df["Review Rating"]=df.groupby("Category")["Review Rating"].transform(lambda x:x.fillna(x.median()))
#For each product category, fill the missing ‘Review Rating’ values with the average (or median) rating of that same category.”
```

```python
df.isnull().sum()
```

```python
df.columns=df.columns.str.lower()#this will convert all heading to lower
df.columns=df.columns.str.replace(' ','_')#this will replace space with _(underscore)
df.columns#we are doing this beacuse whenever we work in sql and pyhton,it will make  easy to write columns name
```

```python
#in previous one , purchase_amount_(usd) come like this so we need to change this
df=df.rename(columns={"purchase_amount_(usd)":"purchase_amount"})
df.columns
```

```python
#create a new columnns as age_group
labels=['young','adult','middle_aged','senior']
df['age_group']=pd.qcut(df['age'],q=4,labels=labels)
```

```python
df[['age','age_group']].head(20)
```

```python
#create a colummn purchase_frequency_days
frequency_mapping={
    'Fortnightly':14,
    'Weekly':7,
    'Monthly':30,
    'Quaterly':90,
    'Annually':365,
    'Bi-Weekly':14,
    'Every 3 Months':90,
    }
df['purchase_frequency_days']=df['frequency_of_purchases'].map(frequency_mapping)


```

```python
df[['frequency_of_purchases','purchase_frequency_days']]
```

```python
(df['discount_applied']==df['promo_code_used']).all()#both columns are same (have same value , means every body who gets discount has used promo_code)so we can delete promo_code_used column to reduce redundancy
```

```python
df=df.drop('promo_code_used',axis=1)#remove the columns , axis one bcoz we have to remove column
```
